```python
import pandas as pd

path = "../data/companies_snapshot.csv"
df = pd.read_csv(path, dtype=str, low_memory=False)
df.columns = [c.strip() for c in df.columns]

df.shape
```

```python
df.columns.tolist()
```

```python
# Check whether dissolved companies actually appear in this file.
df["CompanyStatus"].value_counts(dropna=False)
```

```python
# How many companies have a DissolutionDate filled in, and does it relate to status?
df["DissolutionDate"].notna().sum()
```

```python
# How many company numbers are not standard 8-digit format?
df["CompanyNumber"].str.match(r"^\d{8}$").value_counts()
```

```python
# Looking at the format of company numbers that don't match the standard 8-digit pattern.
df.loc[~df["CompanyNumber"].str.match(r"^\d{8}$"), "CompanyNumber"].str[:2].value_counts().head(15)
```

```python
# Check for names that are purely numbers, and names containing "plc"/"limited" only with no company-specific part.
df[df["CompanyName"].str.match(r"^\d+\s*(PLC|LIMITED|LTD)?$", case=False, na=False)].head(10)
```

```python
# Check missing values in just the columns we plan to use.
cols_needed = [
    "CompanyName", "CompanyNumber", "CompanyCategory", "CompanyStatus",
    "IncorporationDate", "SICCode.SicText_1",
    "Accounts.NextDueDate", "Accounts.LastMadeUpDate", "Accounts.AccountCategory",
    "ConfStmtNextDueDate", "ConfStmtLastMadeUpDate", "Mortgages.NumMortCharges",
]
df[cols_needed].isna().sum()
```

```python
# Do companies missing Accounts.LastMadeUpDate tend to be recently incorporated?
df.loc[df["Accounts.LastMadeUpDate"].isna(), "IncorporationDate"].str[-4:].value_counts().head(10)
```

```python
# Check whether Accounts.NextDueDate missingness relates to distress status.
df.loc[df["Accounts.NextDueDate"].isna(), "CompanyStatus"].value_counts().head(10)
```

```python
# Are these Active companies missing NextDueDate also very recently incorporated?
df.loc[(df["Accounts.NextDueDate"].isna()) & (df["CompanyStatus"] == "Active"), "IncorporationDate"].str[-4:].value_counts().head(10)
```

```python
# Check the accounts category for these active companies with no next due date.
df.loc[(df["Accounts.NextDueDate"].isna()) & (df["CompanyStatus"] == "Active"), "Accounts.AccountCategory"].value_counts()
```

```python
# Check for duplicate company numbers, and look at a few raw date values.
print("Duplicate company numbers:", df["CompanyNumber"].duplicated().sum())
print(df["IncorporationDate"].dropna().sample(5, random_state=1).tolist())
```

```python
# How many failed vs active companies remain after restricting to standard 8-digit companies?
standard = df[df["CompanyNumber"].str.match(r"^\d{8}$")]
standard["CompanyStatus"].value_counts()
```
